```python
def kaprekar_loop(n):
    v = set()
    if n == 6174 or n ==495 or n ==549945 or n ==631764:
        return print (n)
    elif karpekar_check(n)== False:
        return print ('!     ' + str(n) + ',     ')
    else:        
        while True:
            if n not in v:
                print(n)
                v.add(n)
                n = kaprekar_step(numerics(n))
            elif n in [495, 6174, 549945, 631764]:
                break
            elif n in v:
                return print('  - {},   ...'.format(str(n)))
                break

    
def karpekar_check(n): 
    if n <= 100 or n == 1000 or n == 100000 or n >=1000000:
        return False
    elif len(list(set(numerics(n)))) == 1:
        return False
    else:
        return True 

def numerics(n):
    a = []
    for i in str(n):
        a.append(i)
    a = list(map(int, a))
    return a


def kaprekar_step(L):
    max1 = sorted(L)
    max2 = max1[::-1]
    razn = int("".join(map(str, max2))) - int("".join(map(str, max1))) 
    return razn
```
